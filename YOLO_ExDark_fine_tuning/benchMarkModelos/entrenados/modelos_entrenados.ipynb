```python
from ultralytics import YOLO
import gc

# Lista de modelos
modelos = [
    'yolov5nu.pt',
    'yolov8n.pt',
    'yolov9t.pt',
    'yolov10n.pt',
    'yolo11n.pt'
]

# Ruta al archivo de configuraci√≥n del dataset
ruta_dataset = '/home/david/vision_computador/entrega3/yolo/data/data.yaml'

# Funci√≥n para evaluar un modelo y obtener mAP50
def evaluar_modelo(modelo_path, dataset_path, nombre_evaluacion, epocas):
    try:
        modelo = YOLO(modelo_path)
        # resultados = modelo.val(data=dataset_path, split='test', project='resultados_test', name=nombre_evaluacion)
        resultados = modelo.train(data=dataset_path, epochs=epocas, project='resultados_20_epoc', name=nombre_evaluacion, imgsz=500, batch=8)
        print(f"Modelo {modelo_path} evaluado correctamente")
        return resultados
    except Exception as e:
        print(f"Error al evaluar el modelo {modelo_path}: {e}")
        return None
    finally:
        # Liberar memoria
        del modelo
        gc.collect()
```

```python
# Entrenar modelo 0, yolo5nu.pt
# tiempo de entrenamiento: 20 epochs completed in 0.634 hours.
epocas = 20
nombre_evaluacion = str(epocas)+'epocas'+'_'+modelos[0]
resultados = evaluar_modelo(modelos[0], ruta_dataset, nombre_evaluacion, epocas)
```

```python
# Recorremos los directoriso de ntro de resultados _1_epoc
from pathlib import Path
import pandas as pd

# Ruta a la carpeta principal
ruta_principal = Path('./resultados_1_epoc/')

resultados = {}

# Recorrer los subdirectorios
for subdirectorio in ruta_principal.iterdir():
    if subdirectorio.is_dir():
        res_csv = subdirectorio / 'results.csv'
        if res_csv.exists():
            name = subdirectorio.name
            name = name.split('_')[1]
            print(f"Procesando {name}")
            resultados[name] = pd.read_csv(res_csv)
```

```python
# para todos los capos de resultados nos quedamos con el valor m√°s alto y el m√°s bajo
maximos = {}
minimos = {}

for key in resultados.keys():
    maximos[key] = {}
    minimos[key] = {}
    for col in resultados[key].columns:
        if col != 'epoch':
            maximos[key][col] = resultados[key][col].max()
            minimos[key][col] = resultados[key][col].min()

```

```python
import matplotlib.pyplot as plt

metricas_altas = ['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)', 'metrics/mAP50-95(B)']
colores = ['skyblue', 'lightgreen', 'salmon', 'orange']

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

for i, metrica in enumerate(metricas_altas):
    max_precisiones = {modelo: maximos[modelo][metrica] for modelo in maximos}

    # Graficar los m√°ximos de precisi√≥n
    modelos = list(max_precisiones.keys())
    precisiones = list(max_precisiones.values())

    ax = axs[i // 2, i % 2]
    ax.bar(modelos, precisiones, color=colores[i])
    ax.set_xlabel('Modelos')
    ax.set_ylabel(f'M√°xima {metrica.split("/")[1]}')
    ax.set_title(f'M√°xima {metrica.split("/")[1]} de Cada Modelo')
    ax.set_xticks(range(len(modelos)))
    ax.set_xticklabels(modelos, rotation=45)

    # Ajustar la escala del eje y para resaltar las diferencias
    ax.set_ylim(min(precisiones) - 0.01, max(precisiones) + 0.01)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

metricas_bajas = ['train/box_loss', 'val/box_loss', 'train/cls_loss', 'val/cls_loss', 'train/dfl_loss', 'val/dfl_loss']
colores = ['skyblue', 'lightgreen', 'salmon', 'orange', 'purple', 'brown']

fig, axs = plt.subplots(3, 2, figsize=(15, 15))

for i, metrica in enumerate(metricas_bajas):
    min_valores = {modelo: minimos[modelo][metrica] for modelo in minimos}

    # Graficar los m√≠nimos de las m√©tricas
    modelos = list(min_valores.keys())
    valores = list(min_valores.values())

    ax = axs[i // 2, i % 2]
    ax.bar(modelos, valores, color=colores[i])
    ax.set_xlabel('Modelos')
    ax.set_ylabel(f'M√≠nima {metrica}')
    ax.set_title(f'M√≠nima {metrica} de Cada Modelo')
    ax.set_xticks(range(len(modelos)))
    ax.set_xticklabels(modelos, rotation=45)

    # Ajustar la escala del eje y para resaltar las diferencias
    ax.set_ylim(min(valores) - 0.01, max(valores) + 0.01)

plt.tight_layout()
plt.show()
```
