```python
from ultralytics import YOLO
import os
from pathlib import Path

# Cargar el modelo
model = YOLO('./best.pt')

# Directorio de entrada y salida
input_dir = './input/'
output_dir = './output/'
os.makedirs(output_dir, exist_ok=True)

# Procesar cada video en el directorio de entrada
for video_name in os.listdir(input_dir):
    if video_name.endswith(('.mp4', '.avi', '.mov')):
        video_path = os.path.join(input_dir, video_name)
        
        # Realizar la predicción
        results = model.predict(source=video_path, save=True, save_txt=False, conf=0.25)
        
        # Mover el video anotado al directorio de salida
        output_file = f"{Path(video_name).stem}_annotated.mp4"
        annotated_path = os.path.join('runs', 'detect', 'predict', video_name)
        final_output_path = os.path.join(output_dir, output_file)
        
        if os.path.exists(annotated_path):
            os.rename(annotated_path, final_output_path)
```

```python
from ultralytics import YOLO
import os
from pathlib import Path

# Cargar el modelo
model = YOLO('./best.pt')

# Directorio de entrada y salida
input_dir = './input/'
output_dir = './output/'
os.makedirs(output_dir, exist_ok=True)

# Procesar cada imagen en el directorio de entrada
for img_name in os.listdir(input_dir):
    if not img_name.endswith(('.jpg', '.jpeg', '.png')):
        continue
    img_path = os.path.join(input_dir, img_name)
    results = model(img_path)
    
    # Guardar solo la imagen con la detección
    output_path = os.path.join(output_dir, img_name)
    results[0].save(output_path)
```

```python
import imageio
import os
from pathlib import Path

# Leer el video
predict_dir = './runs/detect/predict/'
output_dir = './output/'
# Procesar cada video en el directorio de predicciones
for video_name in os.listdir(predict_dir):
    if video_name.endswith(('.mp4', '.avi', '.mov')):
        video_path = os.path.join(predict_dir, video_name)
        
        # Leer el video
        reader = imageio.get_reader(video_path)

        # Escribir el GIF
        output_gif_path = os.path.join(output_dir, f"{Path(video_name).stem}.gif")
        writer = imageio.get_writer(output_gif_path)

        for frame in reader:
            writer.append_data(frame)

        writer.close()
```
